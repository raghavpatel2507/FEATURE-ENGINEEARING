## Standardization


```python
import seaborn as sns 
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
import numpy as np
```

```python
mean=np.mean(df['total_bill'])
std=np.std(df['total_bill'])
```

```python
mean,std
```

```python
total_bill=list(df['total_bill'])
```

```python
normalized_data=[]
for i in total_bill:
    z_score=(i-mean)/std
    normalized_data.append(z_score)
    
    
print(normalized_data)    
```

```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(normalized_data)
```

```python
sns.distplot(normalized_data)
```

```python
sns.distplot(df['total_bill'])
```
