## imputation technique
### 1-mean value imputation
#### this technique works well when your data is normally distributed

```python
sb.distplot(df['age']) # normally distributed
```

```python
sb.histplot(df['age'])
```

```python
df['age_mean']=df['age'].fillna(df['age'].mean())
```

```python
df[['age_mean','age']]
```

```python
# this technique works well when your data is normally distributed
```

# 2-median value imputatoion
### if you have outliers in dataset use this technique 

```python
df['age_median']=df['age'].fillna(df['age'].median())
```

```python
df[['age_median','age']]
```

# 3-mode value imputation-- categorical data

```python
df[df['embarked'].isnull()]
```

```python
mode=df['embarked'].mode()[0]
```

```python
mode
```

```python
df['mode_embarked']=df['embarked'].fillna(mode)
```

```python
df[['mode_embarked','embarked']]
```

```python
df['mode_embarked'].isnull().sum()
```
